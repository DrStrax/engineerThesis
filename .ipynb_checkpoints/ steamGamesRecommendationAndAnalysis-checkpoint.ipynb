```python
from surprise import AlgoBase
from surprise import evaluate

class updatedSingleValueDecomposition(AlgoBase):

    def __init__(self, factors=60, epochs=20, learningRate=.005, regularization=.02, std = .1, mean = 0):

        self.factors = factors
        self.epochs = epochs
        self.learningRate = learningRate
        self.regularization = regularization
        self.std = std
        self.mean = mean
        AlgoBase.__init__(self)

    def train(self, trainset):

        AlgoBase.train(self, trainset)
        self.sgd(trainset)

    def sgd(self, trainset):

        # user biases
        userBias = np.zeros(trainset.n_users)
        # item biases
        itemBias = np.zeros(trainset.n_items)
        
        userFactors = np.random.normal(self.mean, self.std,(trainset.n_users, self.factors))
        itemFactors = np.random.normal(self.mean, self.std,(trainset.n_items, self.factors))

        user = 0
        item = 0
        rating = 0
        globalMean = self.trainset.global_mean
        err = 0
        r = 0

        learningRate = self.learningRate

        regularization = self.regularization
        f = 0
        dot = 0
        puf = 0
        qif = 0

        for currentEpoch in range(self.epochs):
            for user, item, raiting in trainset.all_ratings():

                # compute current error
                dot = 0  # <q_i, p_u>
                for factor in range(self.factors):
                    dot += itemFactors[item, factor] * userFactors[user, factor]
                err = r - (globalMean + userBias[user] + itemBias[item] + dot)

                # update biases
                userBias[user] += learningRate * (err - regularization * userBias[user])
                itemBias[item] += learningRate * (err - regularization * itemBias[item])

                # update factors
                for f in range(self.factors):
                    puf = userFactors[user, f]
                    qif = itemFactors[item, f]
                    userFactors[user, f] += learningRate * (err * qif - regularization * puf)
                    itemFactors[item, f] += learningRate * (err * puf - regularization * qif)

        self.userBias = userBias
        self.itemBias = itemBias
        self.userFactors = userFactors
        self.itemFactors = itemFactors

    def estimate(self, user, item):

        est = self.trainset.global_mean

        if self.trainset.knows_user(user):
            est += self.userBias[user]

        if self.trainset.knows_item(item):
            est += self.itemBias[item]

        if self.trainset.knows_user(user) and self.trainset.knows_item(item):
            est += np.dot(self.itemFactors[item], self.userFactors[user])

        return est
```

```python
algo = updatedSingleValueDecomposition()

# Evaluate performances of our algorithm on the dataset.
perf = evaluate(algo, data, measures=['RMSE', 'MAE'])

print_perf(perf)
```

```python
from collections import defaultdict

from surprise import Dataset


def get_top_n(predictions, n=10):
    # First map the predictions to each user.
    top_n = defaultdict(list)
    for uid, iid, true_r, est, _ in predictions:
        top_n[uid].append((iid, est))

    # Then sort the predictions for each user and retrieve the k highest ones.
    for uid, user_ratings in top_n.items():
        user_ratings.sort(key=lambda x: x[1], reverse=True)
        top_n[uid] = user_ratings[:n]

    return top_n


reader = Reader(rating_scale=(0, 2000))
trainset = data.build_full_trainset()
algo = updatedSingleValueDecomposition()
algo.train(trainset)

testset = trainset.build_anti_testset()
predictions = algo.test(testset)

top_n = get_top_n(predictions, n=10)

for uid, user_ratings in top_n.items():
    print(uid, [iid for (iid, _) in user_ratings])
```

```python
user_item_matrix = pd.pivot_table(playData, index = 'uID', columns ='game', values = 'hours',fill_value= 0)
user_item_matrix
```

```python
#normalized_df=(user_item_matrix-user_item_matrix.min())/(user_item_matrix.max()-user_item_matrix.min())

def normalize(df):
    result = df.copy()
    for feature_name in df.columns:
        max_value = df[feature_name].max()
        min_value = df[feature_name].min()
        result[feature_name] = 1+ (df[feature_name] - min_value)* 9 / (max_value - min_value)
    return result

user_item_matrix = normalize(user_item_matrix)
```

```python
user_item_matrix = user_item_matrix.reset_index().melt(id_vars='uID',value_name='hours')
```

```python
playDataNormalized = user_item_matrix.ix[~(user_item_matrix['hours'] == 1)]
playDataNormalized['hours'] = playDataNormalized['hours'].values.astype(int)
playDataNormalized
```
